```python
def hollowtriangle(n):
  for i in range(1,n+1):
    for j in range(1,2*n):
      if i==n or j==n-i+1 or j==n+i-1:
       print("*",end=" ")
      else:
       print(" ",end=" ")
    print()
hollowtriangle(5)
```

```python
# method 1
def diamond(n):
  for i in range(1,n+1):
     for j in range(n-i):
       print(" ",end=" ")
     for k in range(1,2*i):
       print("*",end=" ")
     print()
  for i in range(n-1,0,-1):
     for j in range(n-i):
       print(" ",end=" ")
     for k in range(1,2*i):
       print("*",end=" ")
     print()
diamond(10)
```

```python
def hollowdiamond(n):
  for i in range(1,n+1):
     for j in range(n-i):
       print(" ",end=" ")
     for k in range(1,2*i):
      if k==1 or k==2*i-1:
       print("*",end=" ")
      else:
        print(" ",end=" ")
     print()
  for i in range(n-1,0,-1):
     for j in range(n-i):
       print(" ",end=" ")
     for k in range(1,2*i):
      if k==1 or k==2*i-1:
        print("*",end=" ")
      else:
        print(" ",end=" ")
     print()
hollowdiamond(10)
```
